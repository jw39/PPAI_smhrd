```python
# í‰ì  ê¸°ì¤€
# ê¸ì • : 5, 4
# ì¤‘ë¦½ : 3
# ë¶€ì • : 2, 1

data['í‰ì ']

# í‰ì  ê¸°ì¤€
def rating_replace(rating):
    if rating in [5, 4]:  # ê¸ì •ì€ 2ë¡œ ë³€ê²½
        return 2
    elif rating == 3:  # ì¤‘ë¦½ì€ 1ë¡œ ë³€ê²½
        return 1
    else:  # ë¶€ì •ì€ 0ìœ¼ë¡œ ë³€ê²½
        return 0

data['ê°ì„±ë¶„ì„ ê²°ê³¼'] = data['í‰ì '].apply(rating_replace)
data
```

```python
# ë¼ì´ë¸ŒëŸ¬ë¦¬ ë¶ˆëŸ¬ì˜¤ê¸°
from konlpy.tag import Okt
okt = Okt()
```

```python
list_ = []

for i in X:
  list_.append(okt.pos(i, join=True))

list_
```

```python
import pandas as pd

# ëª…ì‚¬ì™€ íƒœê·¸ë¥¼ ì €ì¥í•  ë¦¬ìŠ¤íŠ¸
morph = []
tag = []

# ëª…ì‚¬ë¥¼ ì¶”ì¶œí•˜ëŠ” ê³¼ì •
for sentence in list_:
    for word in sentence:
        word_morph, word_tag = word.split('/')
        if word_tag == 'Noun':  # ëª…ì‚¬ë§Œ ì¶”ì¶œ
            morph.append(word_morph)
            tag.append(word_tag)

# ë°ì´í„°í”„ë ˆì„ìœ¼ë¡œ ë³€í™˜

df = pd.DataFrame({'morph': morph, 'tag': tag})

print(df)
```

```python
def myTokenizer(text):
    d = pd.DataFrame(okt.pos(text), columns=['morph','tag'])
    d.set_index('tag', inplace=True)
    if  ('Noun' in d.index):
        return d.loc[d.index.intersection(['Noun']),'morph'].values
    else :
        return []
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
#TF-IDF Vectorizerì— ì—°ê²°í•´ì£¼ê¸°
tfidf_pos = TfidfVectorizer(tokenizer=myTokenizer)
tfidf_pos.fit(X)
tfidf_pos.vocabulary_
```

```python
# ì‹¤ì œ ë¬¸ì¥ì— ë‹¨ì–´ë¥¼ í† í°ê°’ìœ¼ë¡œ ë³€í™˜í•´ì£¼ëŠ” ì‘ì—… ì§„í–‰ : 8ë¶„ì •ë„ ê±¸ë¦¼
X_trains = tfidf_pos.transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_trains, y, test_size=0.2, random_state=42)
```
