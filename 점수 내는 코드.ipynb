```python
import pandas as pd

# CSV 파일 읽기
file_path = './data/result_review_all_18_final_scores (1).csv'
df = pd.read_csv(file_path)

# 드롭할 컬럼 리스트
columns_to_drop = [ '닉네임', '나이', '피부 타입', '성별','평점', '리뷰내용', '나이 그룹', '감성분석 결과', '실제 평점', '긍정 확률 (%)', '부정 확률 (%)', '중립 확률 (%)']

# 데이터프레임에서 지정한 컬럼들 삭제
score = df.drop(columns=columns_to_drop)

# 결과 출력
score
```

```python
grouped = score.groupby('제품명').agg(
    평균_점수=('점수', 'mean'),
    총_점수=('점수', 'sum')
).reset_index()

# 평균 점수 기준으로 내림차순 정렬
grouped_sorted = grouped.sort_values(by='평균_점수', ascending=False)

# 결과 출력
grouped_sorted.head()
```

```python
# CSV 파일로 저장
output_file_path = './data/product_score.csv'
grouped_sorted.to_csv(output_file_path, index=False)

print(f'CSV 파일이 저장되었습니다: {output_file_path}')
```

```python

# 그룹화 및 집계
grouped = score.groupby('제품명').agg(
    평균_점수=('점수', 'mean'),
    총_점수=('점수', 'sum'),
    리뷰_수=('점수', 'count')
).reset_index()

# 리뷰 수에 따른 가중 평균 점수 계산
# 가중치: 총 점수와 리뷰 수의 조합
grouped['가중_평균_점수'] = grouped['평균_점수'] * (grouped['리뷰_수'] / grouped['리뷰_수'].max())

# 가중 평균 점수 기준으로 내림차순 정렬
grouped_sorted = grouped.sort_values(by='가중_평균_점수', ascending=False)

# 결과 출력
grouped_sorted[['제품명', '가중_평균_점수', '총_점수', '리뷰_수']].head(5)
```
