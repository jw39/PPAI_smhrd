```python
import pandas as pd

df_review = pd.read_csv('./data/df_dropna_review.csv', encoding='utf-8')
# df_review.drop(columns=['Unnamed: 0'], inplace = True)
df_review['sub_review']
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer # TF-IDF 객체
from konlpy.tag import Okt
# 샘플데이터를 활용해 TF-IDF 파악
# 벡터화 수행 전 맛보기로 샘플데이터 생성
sample_text = df_review['sub_review']

# TF-IDF 객체 초기화
sample_tfidf = TfidfVectorizer()
# 단어사전 구축
sample_tfidf.fit(sample_text)
# 단어사전 확인
sample_tfidf.vocabulary_
okt = Okt()
# 확인
okt.morphs(sample_text[0])
# 품사 확인
okt.pos(sample_text[0])
# 명사만 추출
okt.nouns(sample_text[0])
# BOW, TF-IDF는 기본적으로 '공백' 기준으로 토큰화 수행
# 토큰의 단위를 형태소 분석기를 응용해서 설정
# 토큰화 도구 생성( 사용자 정의 함수 )
def my_tokenizer(text) :
    return okt.nouns(text)
tfidf_okt = TfidfVectorizer(tokenizer = my_tokenizer, token_pattern = None)
# 단어사전 구축
tfidf_okt.fit(sample_text)
# 단어사전 확인
tfidf_okt.vocabulary_
# 형태 확인( 컴퓨터 기준 )
tfidf_okt.transform(sample_text)
# 형태 확인( 사람 기준 )
tfidf_okt.transform(sample_text).toarray()
# df로 확인
df = pd.DataFrame([tfidf_okt.vocabulary_.keys(), tfidf_okt.vocabulary_.values()]).T
# 정렬( 컬럼 1 기준, 인덱스 자리에 1번컬럼 )
df = df.sort_values(by = 1).set_index(1).T
# 단어사전들이 나타내는 가중치 확인
print(sample_text)
pd.concat([df, pd.DataFrame(tfidf_okt.transform(sample_text).toarray())])
```

```python
# 쌤 코드
# 라이브러리
from sklearn.pipeline import make_pipeline # 기능 연결 도구
from sklearn.linear_model import LogisticRegression

pipe_model = make_pipeline(
    TfidfVectorizer(
        tokenizer = my_tokenizer, # 사용자 정의 토크나이저 기준 지정
        token_pattern = None, # 기본 세팅 제거
        ngram_range = (1, 2) # 1-gram, 2-gram을 단어사전으로 채택
    ),
    LogisticRegression(
        max_iter = 8000 # 최대 10000번 반복
    )
)

X_train.shape, X_test.shape, y_train.shape, y_test.shape

# 약 10000개 가량 데이터를 벡터화 수행, 학습( 시간 조금 걸림 )
pipe_model.fit(X_train, y_train)
```

```python
# 쌤 코드
# 평가( 시간 조금 걸림 )
# 73.91 % 정확도
pipe_model.score(X_test, y_test)
# 0.6169154228855721
# 0.6034825870646766
```

```python
# 쌤 코드
# 긍정 : 1 , 부정 : 0 예측
# print(pipe_model.predict(['이 영화 너무 재밌는 거 같아요!']))
# print(pipe_model.predict(['노잼']))

# 긍정, 부정에 영향을 끼친 단어 살펴보기
# 토큰화 도구에 접근
tfidf = pipe_model.steps[0][1]

# 모델 접근
logi = pipe_model.steps[1][1]

# 단어사전 키, 벨류값에 접근
# 10000개 문장 실행 -> 51852개의 단어로 나눠짐
df = pd.DataFrame([tfidf.vocabulary_.keys(), tfidf.vocabulary_.values()]).T
df.columns = ['단어', '인덱스']
# 인덱스 컬럼 기준 오름차순 정렬
df.sort_values(by = '인덱스', inplace = True)

# 가중치 확인
# -> 2차원인데 행은 하나밖에 없음, 그냥 0번 인덱스를 가져와서 1차원으로 변환
logi.coef_[0]
df['가중치'] = logi.coef_[0]

# by 생략 가능, 가중치 기준 오름차순 정렬
# 가중치 : 부정적 ( - ) -> 긍정적( + )
df.sort_values('가중치', inplace = True)

bad = df.head(20)
good = df.tail(20)
# 부정, 긍정 20개 단어 합친 변수 -> 시각화에 이용
top20 = pd.concat([bad, good])

# 그래프 그리기
import matplotlib.pyplot as plt
# 한글 지원 폰트 지정
plt.rcParams['font.family'] = 'Malgun Gothic'
# 음수 기호( - ) 사용
plt.rcParams['axes.unicode_minus'] = False

# 그래프 그리기
plt.figure(figsize = (12,6)) # 크기 조절
plt.title("긍/부정 단어 가중치")
plt.xlabel('단어')
plt.ylabel('가중치', rotation = 0) # 기본 90도로 설정, 0도로 설정해 가로로 정렬
plt.xticks(rotation = 60) # 글자 겹쳐서 읽기 힘드니까 60도 기울여 줌

plt.bar(top20['단어'], top20['가중치'])
plt.show()


```

```python
import numpy as np

# 각 단어의 감정 점수를 사전 형태로 정의
sentiment_scores_dict = dict(zip(df['단어'], df['가중치']))

# TF-IDF 벡터라이저에서 단어 목록 추출
feature_names = tfidf.get_feature_names_out()

# 각 단어의 감정 점수를 가져오기
sentiment_scores = np.array([sentiment_scores_dict.get(word, 0) for word in feature_names])
```

```python
def calculate_review_sentiment_scores(reviews, tfidf_vectorizer, sentiment_scores):
    # 리뷰를 TF-IDF 벡터로 변환
    X_reviews = tfidf_vectorizer.transform(reviews)
    
    scores = []
    for review_vector in X_reviews:
        # 리뷰 벡터와 단어 감정 점수의 내적을 계산하여 감정 점수 산출
        score = np.dot(review_vector.toarray(), sentiment_scores)
        scores.append(score[0])
    return scores

# df_review 데이터에서 'review' 컬럼을 추출
reviews = df_review['review']

# 리뷰에 대한 감정 점수 계산
sentiment_scores = calculate_review_sentiment_scores(reviews, tfidf, sentiment_scores)

# 감정 점수를 데이터프레임에 추가
df_review['sentiment_score'] = sentiment_scores

# 점수 결과를 확인
print(df_review[['review', 'sentiment_score']].head())

df_review.to_csv('./data/sentiment_weights.csv', index=False, encoding='utf-8')

```
