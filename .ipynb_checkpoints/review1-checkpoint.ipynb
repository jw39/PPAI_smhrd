```python
import numpy as np
import pandas as pd
# 유저 파일 읽어오기
df_review = pd.read_csv('./data/result_review.csv', encoding='utf-8')
# df_review = df_review.drop(columns = ['review_idx', 'cos_name', 'user_nm', 'rating', 'review'])
# df_user = df_user.set_index('user_nm')
df_review
```

```python
text_train = df_review["review"]
```

```python
tmp = [doc.split(' ') for doc in text_train]
```

```python
token_list = []
for s in tmp :
    token_list += s
```

```python
from collections import Counter # 빈도수 세어주는 도구
counter = Counter(token_list)
```

```python
counter.most_common(40)
```

```python
# 원본 데이터를 저장
df_review.to_csv('./data/original_reviews.csv', index=False, encoding='utf-8-sig')
```

```python
# 결측치를 제거한 데이터 저장
df_dropna_review = df_review.dropna(subset=['review'])
df_dropna_review.to_csv('./data/df_dropna_review.csv', index=False, encoding='utf-8-sig')
```

```python
# 결측치가 제거된 데이터 확인
print(df_dropna_review.head(10))
```

```python
# 결측치 처리된 거 한 번 찍어봄
print(df_dropna_review.info())
```

```python
# 전처리하는 함수 하나 생성. 특수문자, 숫자, 다중 공백을 단일 공백으로 전혼, 한국어는 포함되도록 함
def clean_text(text):
    import re
    text = text.lower()
    # 특수 문자 및 숫자 제거
    text = re.sub(r'[^\w\s가-힣]', '', text)  # 한글도 포함되도록 수정
    # 다중 공백을 단일 공백으로 변환
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# 전처리한 거를 sub_review 라는 컬럼에 추가 후 확인
df_dropna_review['sub_review'] = df_dropna_review['review'].apply(clean_text)
df_dropna_review['sub_review']

# 기존 리뷰와 전처리 후 리뷰 비교
print(df_dropna_review[['review', 'sub_review']].head())
```

```python
# df_dropna_review 데이터프레임을 새로운 CSV 파일로 저장
df_dropna_review.to_csv('./data/df_dropna_review.csv', index=False)
df_dropna_review
```

```python
# sub_review 있는 버전으로 df_review_vader.csv 생성
df_review_vader = pd.read_csv('./data/df_dropna_review.csv', encoding='utf-8')
df_review_vader
```

```python
# vader 감정 분석기 사용 후 감정 점수 열 추가
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# VADER 감정 분석기 초기화
analyzer = SentimentIntensityAnalyzer()

# 텍스트의 감정 점수를 계산하는 함수 정의
def get_sentiment_vader(text):
    # 텍스트의 감정 점수를 계산합니다.
    score = analyzer.polarity_scores(text)
    # 종합 감정 점수(compound)만 반환합니다.
    return score['compound']

# 데이터프레임의 'sub_review' 열에 대해 감정 점수 계산 후 새로운 'sentiment_score' 열 추가
df_review_vader['sentiment_score_vader'] = df_review_vader['sub_review'].apply(get_sentiment_vader)

# 결과 확인해보니 다 0.0 나옴 ;
# 'review', 'sub_review', 'sentiment_score' 열을 포함하여 데이터프레임의 처음 몇 행을 출력
print(df_review_vader[['review', 'sub_review', 'sentiment_score_vader']].head())
```

```python
# sentiment_score_vader 있는 버전으로 vader csv 생성
df_review_vader.to_csv('./data/df_review_vader.csv', index=False)
```

```python
# 데이터프레임의 처음 몇 행을 출력하여 확인하기
df_review_vader.head()
```

```python
# 한국어라 안 되는 것 같아서 영어로 시도,, 개잘나오는 걸 확인할 수 있음
test_texts = [
    "I love this product! It's amazing.",
    "I hate this. It's the worst experience I've ever had.",
    "It's okay, neither great nor terrible."
]

for text in test_texts:
    print(f"Text: {text}")
    print(f"Sentiment Score: {get_sentiment_vader(text)}")
```

```python

# Komoran 형태소 분석기 객체 생성
komoran = Komoran()

def preprocess_text(text):
    try:
        # 형태소 분석을 통해 명사, 형용사, 동사, 부사 추출
        morphs = komoran.pos(text)
        # 명사, 형용사, 동사, 부사만 추출
        filtered_morphs = [word for word, tag in morphs if tag in ['NNG', 'NNP', 'VA', 'VV', 'MAG', 'MAJ']]
        return ' '.join(filtered_morphs)
    except Exception as e:
        print(f"Error processing text: {e}")
        return ""

# CSV 파일을 데이터프레임으로 읽어오기
input_file_path = './data/df_dropna_review.csv'

# 인코딩을 명시적으로 설정하여 파일 읽기
try:
    df = pd.read_csv(input_file_path, encoding='utf-8')
except UnicodeDecodeError:
    try:
        df = pd.read_csv(input_file_path, encoding='latin1')
    except UnicodeDecodeError:
        df = pd.read_csv(input_file_path, encoding='cp949')  # 추가 인코딩 시도

# 형태소 분석 도구 이름
morph_tool_name = "komoran"

# 'sub_review' 열을 형태소 분석하여 새로운 열에 저장
df['preprocessed_review'] = df['sub_review'].apply(preprocess_text)

# 형태소 분석 결과를 CSV 파일로 저장
output_file_path = f'./data/df_review_with_{morph_tool_name}.csv'
df.to_csv(output_file_path, index=False)
```
